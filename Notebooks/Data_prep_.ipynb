### Exponential smoothing (0s)

```python
#Exponential Smoothing
melted_data['smoothing'] = 0.0
alpha = 0.2  # Smoothing factor
# melted_data[melted_data['Morion ID'] == -876866] - to delete , we dont need it 
# melted_data
# result 76440 rows × 13 columns
```

```python
for morion_id, group in melted_data.groupby('morion_id'):
    sales = group['sales'].copy()
    smoothing = sales.copy()
    
    if not sales.empty and sales.iloc[0] != 0.0:
        smoothing.iloc[0] = sales.iloc[0]

    
    if not sales.empty and sales.iloc[0] == 0.0:
        smoothing.iloc[0] = sales[sales != 0.0].iloc[0]
    
    for i in range(1, len(sales)):
        smoothing.iloc[i] = alpha * sales.iloc[i] + (1 - alpha) * smoothing.iloc[i - 1]

    melted_data.loc[group.index, 'smoothing'] = smoothing

# melted_data
# len(melted_data[melted_data['sales'] == 0.0]), len(melted_data[melted_data['smoothing'] == 0.0])
# result 76440 rows × 13 columns
```

```python
strength_regex_map= {
    'Oral Solid Ordinary': re.compile(r"\d+(,\d+)? mg|\d+(,\d+)? %|\d+(,\d+)? mg/ml|\d+(,\d+)? g"), # DONE - OK
    'Topical Dermatological, Haemorrhoidal, External': re.compile(r"\d+(,\d+)? %|\d+ mg/ml|\d+ mg/g|\d+ mg(?=\s)"), # DONE - OK
    'Parenteral Ordinary': re.compile(r"\d+(,\d+)? %|\d+(,\d+)? mg/ml|\d+(,\d+)? mg/g|\d+(,\d+)? mg(?=/\d+)|\d+ mg|\d+(,\d+)? g"), # DONE - OK
    'Oral Liquid Ordinary': re.compile(r"\d+(,\d+)? %|\d+(,\d+)? mg/ml|\d+(,\d+)? mg/g|\d+(,\d+)? mg(?=/\d+)|\d+ mg(?=/\d+ ml)|\d+ mg(?!,)|\d+(,\d+)? mg(?=/dose)|\d+ мг/\d+(,\d+\sг?)"), # checked
    'Rectal systemic': re.compile(r"\d+(,\d+)? mg|\d+(,\d+)? g"), # Done - OK
    'Oral Solid Retard': re.compile(r'\d+(,\d+)? mg|\d+(,\d+)? g'), # DONE - OK
    'Ophthalmic': re.compile(r'\d+(,\d+)? %|\d+(,\d+)? mg/ml|\d+(,\d+)? mg/g|\d+(,\d+)? mg(?=/\d+)|\d+ mg(?=/\d+ ml)|\d+(,\d+)? mg(?=/dose)'), # Done
    'Other systemic': re.compile(r'\d+(,\d+)? mg(/ml|/g)?(?!,)|\d+(,\d+)? %'), # DONE - OK
    'Non-Human Use and Others': re.compile(r'\d+(,\d+)? mg') # DONE - OK
    
}
    
drug_size_map = {
    'Oral Solid Ordinary': None, # no size value - OK
    'Topical Dermatological, Haemorrhoidal, External': re.compile(r'\d+(,\d+)? g(?!/)\b|\d+ ml\b'),
    'Parenteral Ordinary': re.compile(r"(?<!/)\d+(,\d+)? ml|\d+(,\d+)? g"), 
    'Oral Liquid Ordinary': re.compile(r'(?<=\s)\b\d+(,\d+)? ml|(?<!,)\b\d+ g\b'), # DONE- OK
    'Rectal systemic': None,
    'Oral Solid Retard': None, # no size value
    'Ophthalmic': re.compile(r'\d+(,\d+)? ml|\d+ mg(?=\s|,)'), #DONE - OK
    'Other systemic': None, # no size value
    'Non-Human Use and Others': None, # no size value
}



pack_size_patterns = {
    'all': r"(?<=#)\d+"
}

form = ['cap', 'tab', 'spray', 'emulgel','suppository''gel','cream', 'sol', 'pow', 'syrup','efferv','suspens','granules', 'drops', 'plaster']
```

```python
# Define a function to extract concentration using the dictionary of regexes
def extract_strength(description, nfc): 
    match = re.search(strength_regex_map[nfc], description)
    
    if match:
        return match.group(0)
    return ""

def extract_drug_size(description, nfc): 
    match = None
    if drug_size_map.get(nfc, ''):
        match = re.search(drug_size_map[nfc], description)
    
    if match is not None:
        return match.group(0)
    return ""
    
def extract_pack_size(description, nfc):
    match = re.search(pack_size_patterns['all'], description)
    
    if match is not None:
        return match.group(0)
    return 1

def extract_form(description, nfc):
    for f in form:
        if f in description:
            return f

    return ''
```

```python
# Iterate through the DataFrame using itertuples
for row in data.itertuples(index=True):
    description = row.description
    nfc = row.nfc_1 

    strength = extract_strength(description, nfc)
    drug_size = extract_drug_size(description, nfc)
    pack_size = extract_pack_size(description, nfc)
    pack_type = extract_form(description, nfc)
    
    if len(strength) > 1 :
        data.at[row.Index, 'strength_amount'] = strength.split(' ')[0]
        data.at[row.Index, 'strength_unit'] = strength.split(' ')[1]
    
    if len(drug_size) > 1:
        data.at[row.Index, 'drug_size_amount'] = drug_size.split(' ')[0]
        data.at[row.Index, 'drug_size_unit'] = drug_size.split(' ')[1]
        
    data.at[row.Index, 'pack_size'] = pack_size
    data.at[row.Index, 'form'] = pack_type

# data.shape
# result 76440 rows × 19 columns 
data
```

### Encoding of the values 

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
columns_for_encoding = ['atc_code_5', 'brand', 'inn', 'nfc_1', 'strength_unit', 'drug_size_unit', 'form']
```

```python

for col in columns_for_encoding:

    # Extract the 'atc_code_5' column
    col_to_be_encoded = data[col]
    
    # Apply LabelEncoder to the 'atc_code_5' column
    label_encoder = LabelEncoder()
    col_encoded = label_encoder.fit_transform(col_to_be_encoded)
    
    # Add the encoded labels to the dataframe
    data[f'{col}_encoded'] = col_encoded
    
    # Apply MinMaxScaler to the encoded labels
    scaler = MinMaxScaler()
    col_normalized = scaler.fit_transform(data[[f'{col}_encoded']])
    
    # Add the normalized labels to the dataframe
    data[f'{col}_normalized'] = col_normalized
    
# data.shape
# result (76440, 33)
data
```
